```python
from elasticsearch import Elasticsearch
import pandas as pd

# Établir une connexion avec Elasticsearch
es = Elasticsearch("http://localhost:9200")

# Nom de l'index
index_name = "satisfaction_clients"

# Mapping pour l'index
mapping = {
    "mappings": {
        "properties": {
            "Company": {"type": "keyword"},
            "Customer": {"type": "keyword"},
            "Number_review": {"type": "integer"},
            "Language": {"type": "keyword"},
            "Title": {"type": "text"},
            "Date_review": {"type": "date"},
            "Reply": {"type": "text"},
            "Date_reply": {"type": "date"},
            "Rating": {"type": "float"},
            "Status": {"type": "keyword"},
            "Experience": {"type": "text"},
            "Date_experience": {"type": "date"}
        }
    }
}

# Créer l'index avec le mapping
if not es.indices.exists(index_name):
    es.indices.create(index=index_name, body=mapping)

# Importer les données depuis un fichier CSV dans un DataFrame
csv_url = "https://raw.githubusercontent.com/YACI01/repo_test/main/satisfaction_client.csv"
df = pd.read_csv(csv_url)

# Insérer les données dans Elasticsearch
for index, row in df.iterrows():
    es.index(index=index_name, body=row.to_dict(), id=index)

# Effectuer une requête de recherche sur l'index Elasticsearch
query = {
    "query": {
        "match": {
            "Company": "Company1"  # Recherchez les avis pour une entreprise spécifique (ici, "Company1")
        }
    }
}

# Effectuer la recherche
result = es.search(index=index_name, body=query)

# Afficher les résultats
for hit in result["hits"]["hits"]:
    print(hit["_source"])  # Affiche les documents correspondants à la requête
```
