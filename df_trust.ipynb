```python
data_trust_all = []

# Parcourons les 35 pages de Younited
for i in range(1, 36):
    page_url = "https://www.trustpilot.com/review/www.younited-credit.com?page=" + str(i)
    response = requests.get(page_url)
    soup_trust = bs(response.content, 'html.parser')
    reviews = soup_trust.find_all("div", class_="styles_cardWrapper__LcCPA styles_show__HUXRb styles_reviewCard__9HxJJ")

    # Pour chaque bloc de commentaires d'un client, récupérons les données
    for review in reviews:
        # Créons un dictionnaire data
        data = {}

        data["Company"] = soup_trust.find("h1", class_="typography_default__hIMlQ typography_appearance-default__AAY17 title_title__i9V__").span.get_text(strip=True)
        data["Customer"] = review.find("span", class_="typography_heading-xxs__QKBS8 typography_appearance-default__AAY17").text
        data["Review_number"] = review.find("div", class_="styles_consumerExtraDetails__fxS4S").find("span").text
        data["Language"] = review.find("div", class_="typography_body-m__xgxZ_ typography_appearance-subtle__8_H2l styles_detailsIcon__Fo_ua").text
        data["Comment"] = review.find("h2", class_="typography_heading-s__f7029 typography_appearance-default__AAY17").text
        data["Comment Date"] = review.find("div", class_="styles_reviewHeader__iU9Px").time.text
        reply = review.find("div").find("p", class_="typography_body-m__xgxZ_ typography_appearance-default__AAY17 styles_message__shHhX")
        data['Reply'] = reply.text if reply else None
        data['Reply Date'] = review.find("div", class_="styles_content__Hl2Mi").time.text if reply else None
        data["Star"] = review.find("section", class_="styles_reviewContentwrapper__zH_9M").div["data-service-review-rating"]
        type_element = review.find_next("div", class_="typography_body-m__xgxZ_ typography_appearance-subtle__8_H2l styles_detailsIcon__yqwWi")
        data["Type"] = type_element.find("span").text if type_element else None
        data["Experience Date"] = review.find("div", class_="styles_reviewContent__0Q2Tg").text

        # Ajoutons les infos de chaque variable de data dans la liste data_trust_all
        data_trust_all.append(data)

# Transformons la liste de dictionnaires en dataframe
df_younited = pd.DataFrame(data_trust_all)
df_younited
```
